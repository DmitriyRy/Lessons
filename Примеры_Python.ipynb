Удаление данных из DataFrame

```python
import pandas as pd
df=pd.DataFrame({'x':[1,3,2], 'y':[2,4,1]})
```

```python
df
```

```python
# Удаление произвольной строки
# (индекс строки, axis=0 - строка удаляется по вертикальной оси, вдоль индексов)
# inplace=True - изменения применяются к исходному DataFrame.
df.drop(1,axis=0,inplace=True)
```

```python
df
```

```python
# Удаление столбцов
# 'x' - название удаляемого столбца
# axis=1 - удаляется столбец
df.drop('x',axis=1)
```

```python
# другой способ
# 'x' - название удаляемого столбца
del df['x']
```

```python
df
```

Сравнение датафреймов

```python
df1=pd.DataFrame({'x':[1,3,2],'y':[2,4,1]})
df2=pd.DataFrame({'x':[3,1,2],'y':[0,2,2]})
print(df1)
print(df2)
# выводим в ячейках новой таблицы True если ячейки в массиве 1 >= ячейкам во втором
print(df1>=df2)
```

```python
# можно указать несколько условий при сравнении
# используются символы & (и), | (или)

# выведем те значения df1, которые больше 1
df1[df1>1]
```

```python
# выведем те значения df2, которые удовлетворяют двум условиям
df2[(df2<2)&(df2>0)]
```

```python
# выведем те значения df2, которые удовлетворяют одному из двух условий
df2[(df2<1)|(df2>2)]
```

Поиск пустых ячеек

```python
# два метода .isna() и .isnull()
data={'Имя': ['Анна','Борис',None, 'Джон'], 'Возраст':[25,30,28,None]}
df=pd.DataFrame(data)
```

```python
df
```

```python
print (df.isnull())
```

```python
# для фильтрации строк можно использовать функцию .loc
x=df.isnull()
df.isnull().loc[x['Имя'] == True]
```

Действия с DataFrame

```python
# .sum() - сложение данных в столбце датафрейма
df['column_name'].sum()
# .mean() - расчет среднего значения
df['column_name'].mean()
# .median() - медиана по столбцу
df['column_name'].median()
# .std() - стандартное отклонение
df['column_name'].std()
# пропуски данных (NaN) - игнорируются!
```

```python
# .unique() - позволяет определить количество уникальных значений (в столбце или датафрейме)
df['Название столбца'].unique
```
